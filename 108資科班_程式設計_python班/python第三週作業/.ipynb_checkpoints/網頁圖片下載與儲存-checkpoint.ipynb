```python
import bs4, requests, os

my_url = 'http://www.jyic.net'                     # 勁園國際網頁
html = requests.get(my_url)
print("網頁下載中 ...")
html.raise_for_status()                             # 驗證網頁是否下載成功                      
print("網頁下載完成")

myDir = 'my_img'                                
if os.path.exists(myDir) == False:
    os.mkdir(myDir)                                 # 建立資料夾來儲存圖片

objSoup = bs4.BeautifulSoup(html.text, 'lxml')      

imgTag = objSoup.select('img')                      # 搜尋所有圖片檔案
print("搜尋到的圖片數量 = ", len(imgTag))           
if len(imgTag) > 0:                                 # 如果有找到圖片則執行下載與儲存
    for i in imgTag:                    
        imgUrl = i.get('src')                       
        finUrl = my_url+imgUrl                         # 取得圖片在Internet上的路徑
        print(f"{finUrl} 圖片下載中 ... " )
        picture = requests.get(finUrl)              # 下載圖片
        picture.raise_for_status()                  # 驗證圖片是否下載成功
        print("圖片下載成功 ................................ " )

        # 先開啟檔案, 再儲存圖片
        with open(os.path.join(myDir, os.path.basename(imgUrl)), 'wb') as f :
            for my_storage in picture.iter_content(10240):
                f.write(my_storage)
                                  
```
